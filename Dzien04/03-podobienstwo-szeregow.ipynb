```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = np.array([1, 2, 3 ,4, 5, 6, 7, 6, 5, 4])
t2 = np.array([1.2, 1.8, 3 ,4.6, 4.9, 6.3, 7.4, 5.7, 5.2, 4.3])

plt.plot(t1)
plt.plot(t2)
```

```python
THRESHOLD = 0.2

diff_mean = abs(t1.mean()-t2.mean())
diff_std =  abs(t1.std()-t2.std())               
print(diff_mean, diff_std)
                
if diff_mean < THRESHOLD and diff_std < THRESHOLD:
    print("Zgodne")
else:
    print("Niezgodne")
```

```python
from scipy.stats import pearsonr

pearsonr(t1, t2)
```

```python
# metryka RMSE
t1_mean = t1.mean()
rmse1 = np.sqrt( np.sum( (t1-t1_mean)**2 ) / len(t1) )

t2_mean = t2.mean()
rmse2 = np.sqrt( np.sum( (t2-t2_mean)**2 ) / len(t2)  )

abs(rmse1 - rmse2)
```

```python
# porównanie na podstawie energii sygnałów
abs( np.sum(t1**2) - np.sum(t2**2) )
```
