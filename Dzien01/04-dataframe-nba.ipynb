```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/nba.csv", index_col="Name")
df
```

```python
df.isna().sum()
```

```python
# usuwanie wierszy z NaN (wszedzie)
df.dropna(axis=0, how='all', inplace=True )
```

```python
df.isna().sum()
```

```python
# zamien w College NaN na string pusty
df.College = df.College.replace(pd.np.NaN, "")
```

```python
df.Weight /= 2.2
```

```python
def convert_height(s):
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54

convert_height("6-2")
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
df
```

```python
df.to_csv("nba-ok.csv", index_label="nba_index")
```
