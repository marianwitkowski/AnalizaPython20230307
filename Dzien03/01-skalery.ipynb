```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
np.random.seed(1)

df = pd.DataFrame({
    "x1" : np.random.normal(0, 2, 10_000),
    "x2" : np.random.normal(5, 3, 10_000),
    "x3" : np.random.normal(-5, 5, 10_000)
})
```

```python
scaler = StandardScaler()
scaler.fit_transform(df[["x1"]])
```

```python
df1 = pd.DataFrame( StandardScaler().fit_transform(df), columns=['x1','x2','x3'] )
df1
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,6))

ax1.set_title('Przed')
sns.kdeplot(df.x1, ax=ax1)
sns.kdeplot(df.x2, ax=ax1)
sns.kdeplot(df.x3, ax=ax1)

ax2.set_title('Po')
sns.kdeplot(df1.x1, ax=ax2)
sns.kdeplot(df1.x2, ax=ax2)
sns.kdeplot(df1.x3, ax=ax2)
plt.show()
```

```python
np.random.seed(1)

df = pd.DataFrame({
    "x1" : np.random.chisquare(8, 10_000),
    "x2" : np.random.beta(8, 2, 10_000)*40,
    "x3" : np.random.normal(50, 3, 10_000)
})
```

```python
df1 = pd.DataFrame( MinMaxScaler().fit_transform(df), columns=['x1','x2','x3'] )
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,6))

ax1.set_title('Przed')
sns.kdeplot(df.x1, ax=ax1)
sns.kdeplot(df.x2, ax=ax1)
sns.kdeplot(df.x3, ax=ax1)

ax2.set_title('Po')
sns.kdeplot(df1.x1, ax=ax2)
sns.kdeplot(df1.x2, ax=ax2)
sns.kdeplot(df1.x3, ax=ax2)
plt.show()
```

```python
np.random.seed(1)

df = pd.DataFrame({
    "x1" : np.concatenate([ np.random.normal(20,1,1000), np.random.normal(1,1,50) ]),
    "x2" : np.concatenate([ np.random.normal(30,1,1000), np.random.normal(50,1,50) ])
})
```

```python
df1 = pd.DataFrame( RobustScaler().fit_transform(df), columns=['x1','x2'] )
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(8,6))

ax1.set_title('Przed')
sns.kdeplot(df.x1, ax=ax1)
sns.kdeplot(df.x2, ax=ax1)

ax2.set_title('Po')
sns.kdeplot(df1.x1, ax=ax2)
sns.kdeplot(df1.x2, ax=ax2)
plt.show()
```
